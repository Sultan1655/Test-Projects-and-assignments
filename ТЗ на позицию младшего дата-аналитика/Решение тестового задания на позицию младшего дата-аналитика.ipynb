Задача 1: Выявление аномалий в данных о посещениях сайта

```python
import pandas as pd

# Данные
data = {
    'date': ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05',
             '2024-10-06', '2024-10-07', '2024-10-08', '2024-10-09', '2024-10-10'],
    'unique_visitors': [200, 220, 250, 240, 10000, 230, 245, 240, 235, 50]
}
df = pd.DataFrame(data)

# Определение выбросов с помощью межквартильного размаха
Q1 = df['unique_visitors'].quantile(0.25)
Q3 = df['unique_visitors'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Поиск выбросов
outliers = df[(df['unique_visitors'] < lower_bound) | (df['unique_visitors'] > upper_bound)]
print("Выбросы:")
print(outliers)
```

Обоснование аномалий:

1) Выброс на 2024-10-05 может указывать на техническую ошибку, рекламную кампанию или случайное увеличение посещений.

2) Значение на 2024-10-10 также низкое, что может указывать на технические проблемы с данными.

Задача 2: SQL-запрос для анализа активности пользователей

```python
#Запрос для средней продолжительности сессии каждого пользователя
SELECT user_id, AVG(session_duration) AS avg_session_duration
FROM sessions
GROUP BY user_id;

#Запрос для пользователей, которые провели более 100 минут за день
SELECT user_id, DATE(session_start) AS day, SUM(session_duration) AS total_duration
FROM sessions
GROUP BY user_id, DATE(session_start)
HAVING total_duration > 100;
```

Задача 4: Обработка данных о приложении

```python
# Пример данных
data_app = {
    'session_id': [1, 2, 3, 4],
    'user_id': [100, 101, 102, 103],
    'platform': ['iOS', 'Android', 'iOS', 'Android'],
    'session_duration': [45, 30, 60, 15]
}
df_app = pd.DataFrame(data_app)

# Количество уникальных пользователей по платформам
unique_users = df_app.groupby('platform')['user_id'].nunique()

# Средняя продолжительность сессии для каждой платформы
avg_session_duration = df_app.groupby('platform')['session_duration'].mean()

print("Количество уникальных пользователей по платформам:\n", unique_users)
print("Средняя продолжительность сессии по платформам:\n", avg_session_duration)
```
