## 1. Обработка датафреймов для последующего анализа


### 1.1. Импорт необходимых библиотек

```python
import os
import pandas as pd
from itertools import combinations
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### 1.2. Загрузка CSV-файлов из архива

```python
# Укажите путь к папке с CSV-файлами
data_folder = 'Тестовое задание/датафреймы конструкций'

# Получение списка всех CSV-файлов в папке
csv_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

print(f"Найдено {len(csv_files)} CSV-файлов.")
```

### 1.3. Чтение CSV-файлов в словарь датафреймов

```python
# Словарь для хранения датафреймов
df_dict = {}

for file in csv_files:
    file_path = os.path.join(data_folder, file)
    # Читаем CSV без заголовка и с указанием кодировки, если необходимо
    df = pd.read_csv(file_path, header=None, names=['MAC'], encoding='utf-8')
    # Убираем возможные пробелы и приводим к верхнему регистру
    df['MAC'] = df['MAC'].str.strip().str.upper()
    # Сохраняем в словарь без расширения файла
    key = os.path.splitext(file)[0]
    df_dict[key] = df

print("Данные успешно загружены и обработаны.")
```

## 2. Создание таблицы пересечения уникальных данных между парами датафреймов

### 2.1. Получение уникальных MAC-адресов для каждого датафрейма


```python
# Создаем словарь уникальных MAC-адресов
unique_mac_dict = {name: set(df['MAC']) for name, df in df_dict.items()}
```

### 2.2. Вычисление пересечений между каждой парой датафреймов


```python
# Получаем все возможные пары датафреймов
pairs = combinations(unique_mac_dict.keys(), 2)

# Создаем список для хранения результатов
intersection_results = []

for df1, df2 in pairs:
    set1 = unique_mac_dict[df1]
    set2 = unique_mac_dict[df2]
    intersection = set1.intersection(set2)
    percent = (len(intersection) / len(set1.union(set2))) * 100 if len(set1.union(set2)) > 0 else 0
    intersection_results.append({
        'DataFrame 1': df1,
        'DataFrame 2': df2,
        'Intersection (%)': round(percent, 2)
    })

# Создаем итоговый датафрейм
intersection_df = pd.DataFrame(intersection_results)

print("Таблица пересечений между датафреймами создана.")
intersection_df.head()
```

### 2.3. Сохранение таблицы пересечений в CSV

```python
intersection_df.to_csv('intersection_table.csv', index=False, encoding='utf-8-sig')
print("Таблица пересечений сохранена как 'intersection_table.csv'.")
```

## 3. Создание таблицы количества повторов внутри каждого датафрейма

### 3.1. Функция для расчета повторов


```python
def calculate_repeats(df):
    counts = df['MAC'].value_counts()
    average = counts.mean()
    one_plus = (counts >= 1).sum()
    three_plus = (counts >= 3).sum()
    five_plus = (counts >= 5).sum()
    return pd.Series({
        'Average Repeats': round(average, 2),
        '1+ Repeats': one_plus,
        '3+ Repeats': three_plus,
        '5+ Repeats': five_plus
    })
```

### 3.2. Применение функции к каждому датафрейму


```python
# Создаем список для хранения результатов
repeats_results = []

for name, df in df_dict.items():
    repeats = calculate_repeats(df)
    repeats_results.append({
        'DataFrame': name,
        'Average Repeats': repeats['Average Repeats'],
        '1+ Repeats': repeats['1+ Repeats'],
        '3+ Repeats': repeats['3+ Repeats'],
        '5+ Repeats': repeats['5+ Repeats']
    })

# Создаем итоговый датафрейм
repeats_df = pd.DataFrame(repeats_results)

print("Таблица повторов внутри датафреймов создана.")
repeats_df.head()
```

### 3.3. Сохранение таблицы повторов в CSV


```python
repeats_df.to_csv('repeats_table.csv', index=False, encoding='utf-8-sig')
print("Таблица повторов сохранена как 'repeats_table.csv'.")
```

## 4. Подготовка отчета с помощью инструментов визуализации в Python

### 4.1. Визуализация таблицы пересечений


```python
# Создаем матрицу пересечений
pivot_table = intersection_df.pivot(index='DataFrame 1', columns='DataFrame 2', values='Intersection (%)')

# Заполняем пропущенные значения (набор сам с собой)
for df in unique_mac_dict.keys():
    if df not in pivot_table.index:
        continue
    pivot_table.loc[df, df] = 100.0

# Заполняем пропущенные значения нулями
pivot_table = pivot_table.fillna(0)


# Настройка размера графика
plt.figure(figsize=(40, 40))

# Создание тепловой карты
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap='YlGnBu')

plt.title('Процент пересечения уникальных MAC-адресов между датафреймами')
plt.ylabel('DataFrame 1')
plt.xlabel('DataFrame 2')
plt.tight_layout()
plt.savefig('intersection_heatmap.png')
plt.show()

print("Тепловая карта пересечений сохранена как 'intersection_heatmap.png'.")
```

### 4.2. Визуализация таблицы повторов

Используем гистограммы для отображения количества повторов.


```python
# Установка стиля
sns.set(style="whitegrid")

# Фигурная сетка для нескольких графиков
fig, axes = plt.subplots(2, 2, figsize=(40, 40))
fig.suptitle('Количество повторов MAC-адресов внутри датафреймов', fontsize=16)

# Среднее количество повторов
sns.barplot(x='DataFrame', y='Average Repeats', data=repeats_df, ax=axes[0, 0])
axes[0, 0].set_title('Среднее количество повторов')
axes[0, 0].tick_params(axis='x', rotation=90)

# 1+ повторы
sns.barplot(x='DataFrame', y='1+ Repeats', data=repeats_df, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('MAC-адреса с 1+ повторами')
axes[0, 1].tick_params(axis='x', rotation=90)

# 3+ повторы
sns.barplot(x='DataFrame', y='3+ Repeats', data=repeats_df, ax=axes[1, 0], palette='magma')
axes[1, 0].set_title('MAC-адреса с 3+ повторами')
axes[1, 0].tick_params(axis='x', rotation=90)

# 5+ повторы
sns.barplot(x='DataFrame', y='5+ Repeats', data=repeats_df, ax=axes[1, 1], palette='cividis')
axes[1, 1].set_title('MAC-адреса с 5+ повторами')
axes[1, 1].tick_params(axis='x', rotation=90)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('repeats_barplots.png')
plt.show()

print("Гистограммы повторов сохранены как 'repeats_barplots.png'.")
```
